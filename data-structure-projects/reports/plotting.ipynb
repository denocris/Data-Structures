```python
import numpy as np
#from mpl_toolkits.mplot3d import Axes3D
from pylab import *
```

```python
def time(filename, sort_type):
    f = open(filename,"r")

    ff = [line.strip().split('\t') for line in f]
    ff = [x[0] for x in ff ]
    col0 = [float(y.split()[0]) for y in ff]
    col1 = [int(y.split()[1]) for y in ff]
    col2 = [float(y.split()[2]) for y in ff]
    #print col2
    size = [2000, 5000, 10000, 20000, 50000, 100000, 20000]

    unsort = [col2[i] for i in range(0,len(col2),3)]
    sort = [col2[i] for i in range(1,len(col2),3)]
    mosort = [col2[i] for i in range(2,len(col2),3)]
    if sort_type == "unsort":
        ret = unsort
    elif sort_type == "sort":
        ret = sort
    else:
        ret = mosort
    return ret


size = [2000,5000,10000,20000,50000,100000,200000]
```

```python
plt.figure()
plt.title('Time Already Sorted')
#plt.plot(size, time("simple.md", "sort"), 'b', label='simple')
plt.plot(size, time("bubble.md", "sort"), 'r',linestyle='--', label='bubble')
plt.plot(size, time("insertion.md", "sort"), 'm',linestyle='--',  label='insertion')
plt.plot(size, time("quick.md", "sort"), 'g',linestyle='--',  label='quick')
plt.plot(size, time("merge.md", "sort"), 'b',linestyle='--',  label='merge')
plt.plot(size, time("hybrid.md", "sort"), 'y',linestyle='--',  label='hybrid')
plt.xlabel('size')
plt.ylabel('time(s)')
plt.legend(loc='upper left')
rcParams['figure.figsize'] = (10, 6)
plt.show()
    
```

```python
plt.figure()
plt.title('Time Almost Sorted')
plt.plot(size, time("simple.md", "alsort"), 'r',linestyle='-',  label='simple')
#plt.plot(size, time("bubble.md", "alsort"), 'r', label='bubble')
plt.plot(size, time("insertion.md", "alsort"), 'm', linestyle='--', label='insertion')
plt.plot(size, time("quick.md", "alsort"), 'g', linestyle='--',  label='quick')
plt.plot(size, time("merge.md", "alsort"), 'b', linestyle='--',  label='merge')
plt.plot(size, time("hybrid.md", "alsort"), 'y', linestyle='--', label='hybrid')
plt.xlabel('size')
plt.ylabel('time(s)')
plt.legend(loc='lower left')
rcParams['figure.figsize'] = (10, 6)
plt.show()
```
